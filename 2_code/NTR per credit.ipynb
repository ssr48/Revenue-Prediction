```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.width', 1000)
import seaborn as sns
```

```python
df = pd.read_excel('../3_res/ntr.xlsx')
```

```python
df.head()
```

```python
z_i = df.pivot_table(index=['TERM', 'CRS_ID'], aggfunc='size')
z_i = pd.DataFrame(z_i)
z_i = z_i[z_i[0]>1]
z_i.reset_index(inplace=True)
z_i['dups']=1
z_i.drop(0, axis=1,inplace=True)
z_i
```

```python
df = df.merge(z_i, on=['TERM', 'CRS_ID'], how='left')
# df[df['dups']==1] #check to ascertain the credit hours are distributed appropriately
```

```python
df.head()
```

```python
df['TERM'].value_counts(dropna=False)
```

```python
df['FY']=''
df['FY'][(df['TERM']=='2019F')|(df['TERM']=='2020S')]='FY20'
df['FY'][(df['TERM']=='2018F')|(df['TERM']=='2019S')]='FY19'
df['FY'][(df['TERM']=='2017F')|(df['TERM']=='2018S')]='FY18'
df['FY'].value_counts(dropna=False)
```

```python
df2 = df.pivot_table(index=['FY', 'FDB_College', 'FDB_Dept'], values = ['CRS_AT_HRS', 'NET_CRS_REVENUE'], aggfunc=np.sum)
df2.head()
```

```python
df2['NTR/Credit']=df2['NET_CRS_REVENUE']/df2['CRS_AT_HRS']
df2.head()
```

```python
df2.reset_index(inplace=True)
df2.head()
```

```python
df2.to_excel('../3_res/ntr_per_credit.xlsx', index=False)
```
